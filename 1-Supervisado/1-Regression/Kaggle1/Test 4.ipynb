### Selecciono las columnas definitivas. 


```python
# Quito 'Producto'
test_1 = test.loc[:,['Company','Cpu','Inches','Ram','TypeName','Weight','Memory','ScreenResolution', 'Gpu', 'OpSys']]
```

```python
test_1.info()
```

```python
# En ScreenResolution, multiplico los ejes de la pantalla.  
# Antes de eso, trabajo sobre el string, quedandome con los 9 ultimos digitos, los separo por el 'x', me quedan ancho x alto separados, 
# y mediante una iteracion los multiplico.

test_1['ScreenResolution'] = test_1['ScreenResolution'].str[-9:].str.replace(' ','').str.split('x').apply(lambda h: np.prod([int(i) for i in h]))

test_1['ScreenResolution'].unique()
```

```python
# Limpio los strings de Memory hasta tener un numero unico (si hay dos, sumados)

test_1['Ram'] = test_1['Ram'].str.replace('GB','').astype(int)
```

```python
test_1['Weight'] = test_1['Weight'].str.replace('kg','').astype(float)
```

```python
test_1['Memory'] = test_1['Memory'].str.replace('GB HDD','')
test_1['Memory'] = test_1['Memory'].str.replace('GB SSD','')
test_1['Memory'] = test_1['Memory'].str.replace('TB HDD','024')
test_1['Memory'] = test_1['Memory'].str.replace('TB SSD','024')
test_1['Memory'] = test_1['Memory'].str.replace('.0TB Hybrid','024')
test_1['Memory'] = test_1['Memory'].str.replace(' ','')
test_1['Memory'] = test_1['Memory'].str.replace('GBFlashStorage', '')
test_1['Memory'] = test_1['Memory'].str.replace('GBHybrid', '')
test_1['Memory'] = test_1['Memory'].str.replace('1.0024', '1024')
```

```python
test_1['Memory'] = test_1['Memory'].str.split('+').apply(lambda x: sum([int(i) for i in x]))
```

```python

test_1['Memory'].unique()
```

```python
# Ahora extraigo los procesadores
test_1['Cpu_prosesor'] = test_1['Cpu'].str.extract('(i3|i5|i7|Celeron|A4|Cortex|A6|A7|A8|A9|A10|A12|Pentium|Core M|Atom|E-Series|Ryzen|Xeon|FX)', expand=False)

# test_1['Cpu_prosesor_num'] = test_1['Cpu_prosesor'].astype('category').cat.codes
```

```python
test_1['Cpu_prosesor'].unique()
```

```python
# Importo los diccionarios con los valores numericos de las variables categoricas. 

import json

with open('Company_dict.json', 'r') as f:
    contenido = f.read()

Company_dict = json.loads(contenido)

with open('NameType_dict.json', 'r') as f:
    contenido = f.read()

NameType_dict = json.loads(contenido)

with open('Processor_dict.json', 'r') as f:
    contenido = f.read()

Processor_dict = json.loads(contenido)

with open('OpSys_dict.json', 'r') as f:
    contenido = f.read()

OpSys_dict = json.loads(contenido)

with open('Gpu_dict.json', 'r') as f:
    contenido = f.read()

Gpu_dict = json.loads(contenido)
```

```python
# Creo las columnas con los nuevos valores

test_1['Company_encoded'] = test_1['Company'].map(Company_dict)
test_1['TypeName_encoded'] = test_1['TypeName'].map(NameType_dict)
test_1['Cpu_prosesor_encoded'] = test_1['Cpu_prosesor'].map(Processor_dict)
test_1['Gpu_encoded'] = test_1['Gpu'].map(Gpu_dict)
test_1['OpSys_encoded'] = test_1['OpSys'].map(OpSys_dict)
```

```python
# Tengo dos valores NaN, entonces los imputo con la media de la columna.

test_1['Cpu_prosesor_encoded'].fillna(test_1['Cpu_prosesor_encoded'].mean(), inplace=True)
test_1['Gpu_encoded'].fillna(test_1['Gpu_encoded'].mean(), inplace=True)
```
